```python
# Sentiment analysis is also text classification.

classifier = pipeline('sentiment-analysis')
result= classifier("I was very glad to see you today.")
print(result)
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, AutoModelForTokenClassification, DistilBertTokenizer, DistilBertForTokenClassification, DistilBertForSequenceClassification
from transformers import pipeline


model_name1 = "distilbert-base-uncased-finetuned-conll03-english"
model_name2 = "distilbert-base-uncased-finetuned-sst-2-english"

mytokenizer1 = DistilBertTokenizer.from_pretrained(model_name2)
mymodel1 = DistilBertForTokenClassification.from_pretrained(model_name2)

classifier = pipeline('sentiment-analysis')
res = classifier("I just finished advanced level. what can i do in the University ?")
print(res)
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer, AutoModelForTokenClassification, DistilBertTokenizer, DistilBertForTokenClassification, DistilBertForSequenceClassification
from transformers import pipeline

model_name3 = "nlptown/bert-base-multilingual-uncased-sentiment"
mymodel3 = AutoModelForSequenceClassification.from_pretrained(model_name3)
mytokenizer3 = AutoTokenizer.from_pretrained(model_name3)

classifier = pipeline('sentiment-analysis', model = mymodel3, tokenizer = mytokenizer3)
res = classifier("I just finished advanced level. what can i do in the University ?")
print(res)
```

```python
from transformers import AutoTokenizer

# Load a pre-trained tokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

# Example text.
text = "I was so not happy with the Barbie Movie"

# Tokenize the text.
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)


# Convert tokens to input IDs
input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs:", input_ids)

# Encode the text
encoded_text = tokenizer.encode(text)
print("Encoded Text:", encoded_text)
```

```python
# Convert the tokens in inputs IDs
input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs:", input_ids)
```

```python
# Encode the text (tokenization + converting the input IDs)
encoded_input = tokenizer(text)
print("Encoded Input:", encoded_input)
```

```python
# Decode the text.
decoded_text = tokenizer.decode(encoded_text)
print("Decoded Text:", decoded_text)
```
