## NLP Task

### Sentiment Analysis

```python
# Sentiment analysis is also text classification.

classifier = pipeline('sentiment-analysis')
result= classifier("I was very glad to see you today.")
print(result)
```

```python
classifier = pipeline('sentiment-analysis', model = 'facebook/bart-large-mnli')
result = classifier('I was very glad to see you today.')
print(result)
```

## Batch Sentiment Analysis

```python
classifier = pipeline(task = 'sentiment-analysis')

task_list = [
    "I really like Autoencoder, they are great", \
    "I am not sure if we can actually evaluate LLMs", \
    "I hate long meetings"

]
classifier(task_list)
```

## Text Generation

```python
# Use a pipeline as a high-level helper.

from transformers import pipeline

text_generator = pipeline('text-generation', model = 'inclusionai/ring-1t')
generated_text = text_generator('Today is a great day in Bamenda')
print("Generated_text: \n",  generated_text[0]['generated_text'])
```

## Question Answering

```python
qa_model = pipeline('question-answering')
question= "I just finished advanced level. what can i do in the University ?"
context = "You are a Z-Learn helpful assistance to guide student through thier career paths."
qa_model(question = question, context = context)
```
