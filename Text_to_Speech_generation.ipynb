```python
from transformers import pipeline
```

```python
text = "Python is a high-level, general purpose programming language"

pip = pipeline("text-to-speech", model="suno/bark-small", device="cuda")
output = pip(text)
```

```python
output
```

```python
# Convert our array into Audio.
from IPython.display import Audio
Audio(output["audio"], autoplay=True, rate=output['sampling_rate'])
```
